```python
# Cleaning Raw Data from DJI and MIB
# By Chris Chiang
import pandas as pd
import datetime as dt
import numpy as np
```

```python
# Load in raw data
data_loc = '../raw_data/Market_data/'
dow_df = pd.read_csv(
    data_loc + 'Dow Jones Industrial Average Historical Data.csv')
MIB_df = pd.read_csv(data_loc + 'FTSE MIB Historical Data.csv')
dow_df.head()
```

```python
# Grab only change, price, and reformat date for DOW
new_dow_df = pd.DataFrame()
for index, row in dow_df.iterrows():
    date = dt.datetime.strptime(dow_df.iloc[index]['Date'], '%b %d, %Y').date()
    row = pd.DataFrame(
        {'Date': date, 'Price': row['Price'], 'Change %': row['Change %']}, index=[index])
    new_dow_df = pd.concat([row, new_dow_df])
new_dow_df.tail()
```

```python
# Grab only change, price, and reformat date for MIB
new_MIB_df = pd.DataFrame()
for index, row in MIB_df.iterrows():
    date = dt.datetime.strptime(MIB_df.iloc[index]['Date'], '%b %d, %Y').date()
    row = pd.DataFrame(
        {'Date': date, 'Price': row['Price'], 'Change %': row['Change %']}, index=[index])
    new_MIB_df = pd.concat([row, new_MIB_df])
new_MIB_df.tail()
```

```python
new_dow_df.to_csv('../cleaned_data/us_market.csv', index=False)
new_MIB_df.to_csv('../cleaned_data/italy_market.csv', index=False)
```
