```python
# Cleaning Raw Data from Johns Hopkins
# By Chris Chiang
import pandas as pd
import datetime as dt
import numpy as np
```

```python
data_loc = '../raw_data/COVID-19/csse_covid_19_data/csse_covid_19_daily_reports/'
# Setting up dates
start_date = dt.date(2020, 1, 22)
# last_date = dt.date(2020, 1, 24)
last_date = dt.date(2020, 5, 6)
# last_date = dt.date.today() - dt.timedelta(days=1)  
# country format changed for the last 6 days of data so need to adjust
days = (last_date - start_date).days
```

```python
# forloop init
current_date = start_date
combine_df = pd.DataFrame()
# i is the index of combine_df
cur_index = 0
for day in range(round(int(days)/7)):
    # convert date into string and load in csv
    load_string = current_date.strftime("%m-%d-%Y")
    try:
        today_df = pd.read_csv(data_loc + load_string +
                               '.csv')
    except:
        print('no data')
        print(current_date)
    try:
        stateRow = today_df['Province/State']
        countryRow = today_df['Country/Region']
    except:
        # deal with data column format change
        stateRow = today_df['Province_State']
        countryRow = today_df['Country_Region']
    # total count init
    current_df=pd.DataFrame()
    current_df['Province/State']=stateRow
    current_df['Country/Region']=countryRow
    current_df['Confirmed']=today_df['Confirmed'].fillna(0)
    current_df['Deaths']=today_df['Deaths'].fillna(0)
    current_df['Date']=current_date
    try:
        current_df['Latitude']=today_df['Latitude']
        current_df['Longitude']=today_df['Longitude']
    except:
        try:
            current_df['Latitude']=today_df['Lat']
            current_df['Longitude']=today_df['Long_']
        except:
            current_df['Latitude']=np.nan
            current_df['Longitude']=np.nan
    # combine_df=combine_df.append(current_df)
    current_df.to_csv('../cleaned_data/In_progress/covid_weekly/covid_weekly'+load_string+'.csv', index=False)
    current_date = current_date + dt.timedelta(days=7)
# combine_df
```
